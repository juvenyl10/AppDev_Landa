# **Exercise 2:**

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours.

```python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

if hours > 40:
  overtime = (hours - 40) * 1.5
  pay = (40 + overtime) * rate
  print(f"Pay: {pay}")
else:
  pay = hours * rate
  print(f"Pay: {pay}")
```

# **CHALLENGE**

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

```python
Score            Grade
[ 97, 100]        1.0
[ 93,  97)        1.25
[ 89,  93)        1.5
[ 85,  89)        1.75
[ 81,  85)        2.0
[ 77,  81)        2.25
[ 74,  77)        2.5
[ 0,   74)        3.0
```

```python
score = int(input("Enter score between 0 and 100: "))
if score >= 0 and score <= 100:
   print("Score\t\t\tGrade")
   if score >= 97:
        print("[97, 100]\t\t1.0")
   elif score >= 93:
        print("[93, 97)\t\t1.25")
   elif score >= 89:
        print("[89, 93)\t\t1.5")
   elif score >= 85:
        print("[85, 89)\t\t1.75")
   elif score >= 81:
        print("[81, 85)\t\t2.0")
   elif score >= 77:
        print("[77, 81)\t\t2.25")
   elif score >= 74:
        print("[74, 77)\t\t2.5")
   else:
        print("[0, 74)\t\t\t3.0")
else:
   print("Please enter score between 0 to 100")
```
