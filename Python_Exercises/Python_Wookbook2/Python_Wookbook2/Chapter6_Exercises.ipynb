# **Exercise 1**

Given the code below

```python
text = 'Business Analytics'
```

What will the following expressions produce?

```python
text[4]
text[2:4]
text[:4]
text[4:]
text[-2]
text[-4:-2]
text[-2:]
text[:-2]
```

The program below demonstrates how to slice a string using various indices supplied within square brackets, allowing the extraction of certain substrings from the original string.

```python
text = 'Business Analytics'

print(text[4])
print(text[2:4])
print(text[:4])
print(text[4:])
print(text[-2])
print(text[-4:-2])
print(text[-2:])
print(text[:-2])
```

# **CHALLENGE**


Write a program which repeatedly reads a full name, then does the following:

strips the name of any trailing spaces
capitalizes the first letter
puts the rest into smaller letters

```python
while True:
  name = input("Enter full name: ")
  name = name.strip()
  name = name.capitalize()
  print(name)
  if (name == ""):
    break
```
