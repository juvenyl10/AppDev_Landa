What are the constants in the math module?
- The constants in the math module is:
      1. math.pi for pi.
      2. math.e for logarithm
      3. math.tau for the 2pi

What does math.exp do?
- The math.exp(x) function returns the exponential of x, which is often written as e^x. So, math.exp(x) computes the value of e raised to the power of x.

What about math.pow?
- The math.pow(x, y) is a function that returns x raised to the power of y. math.pow() can take two arguments: x and y, where x is the base and y is the exponent. The result is x ^ y.

How do you calculate logarithms?
- To calculate logarithms math.log() function is use in Python. math.log() function is used to compute the natural logarithm (base e) of a number.

Logarithms base 10?
- Below is the code to calculate logarithms base 10

```python
  import math

  result = math.log10(100)
  print(result)
```

Natural logarithms?
- Natural logarithms are logarithms with base e. Below is the code to calculate the natural logarithms

```python
import math

result = math.log(10)
print(result)
```
